### Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Read and display image

```python
img = cv2.imread(r"C:\Users\ragin\Downloads\wallpapertip_wallpaper-of-girl-with_251848.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.show()
```

### Convert image into Gray format

```python
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(5,5))
plt.imshow(img_gray,cmap='gray')
```

### Gray to inverted image

```python
inverted_image = cv2.bitwise_not(img_gray)
plt.figure(figsize=(5,5))
plt.imshow(inverted_image)
plt.show()
```

### Blurred image

```python
#img_sooth=cv2.GaussianBlur(img_invert,(23,23),sigmaX=0,sigmaY=0)
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.figure(figsize=(5,5))
plt.imshow(blurred_image,cmap='gray')
plt.axis("off")
plt.show()
```

### Blured to inverted blurred

```python
inverted_blurred = cv2.bitwise_not(blurred_image)
plt.figure(figsize=(5,5))
plt.imshow(inverted_blurred,cmap='gray')

plt.show()
```

### Final Sketch image

```python
pencil_sketch = cv2.divide(img_gray, inverted_blurred, scale=230.0)
plt.figure(figsize=(5,5))
plt.imshow( pencil_sketch,cmap='gray')
plt.show()



```

### To Save image

```python
cv2.imwrite('pencil_sketch.jpg', pencil_sketch)
```
